### Correlation and p-value for Attrition vs other variables using pearsonr()
Correlation analysis is used to find the interdependency of variables.

```python
from scipy.stats import pearsonr
```

```python
df.columns
```

### Attrition and Age
- H0 = There is no significant correlation between Attrition and Age
- Ha = There is significant correlation between Attrition and Age 

```python
r, p = pearsonr(df.Attrition, df.Age)
print(r, p)
if p < 0.05:
    print("H0 is rejected")
    print("Ha is accepted")
else:
    print("H0 is accepted")
    print("Ha is rejected")
```

### Inference 
- As we can see p < 0.05 Ha is accepted
- r = -0.15, there is low negative correlation between Attrition and Age
